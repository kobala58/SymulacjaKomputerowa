```python
import random
# import numpy as np
import matplotlib.pyplot as plt
# from matplotlib.pyplot import figure
```

```python
# CONSTANS
SIZE = 50
WINDSIZE = int(0.2 * SIZE)
PHOTORADIUS = 3
OVERLAP = 0.5
```

```python
beg = random.randint(0, SIZE-WINDSIZE)
direction = random.choice(["<", ">"])
windy_cords = []
def wind_vals_gen(x,y):
    if beg<=y<=(beg+WINDSIZE):
        windy_cords.append([x,y])
        return 1
    else:
        return 0
```

```python
points = []
for x in range(SIZE):
    x_tmp = []
    for y in range(SIZE):
        x_tmp.append([x,y, wind_vals_gen(x,y)])
    points.append(x_tmp)
```

```python
# figure()
fig = plt.figure(figsize=(12, 12), dpi=100)
ax1 = fig.add_subplot()
for x in points:
    for y in x:
        if y[-1] != 0:
            ax1.plot(y[0],y[1], color="green", marker=direction)
        else:
            ax1.plot(y[0], y[1], marker="o", color="gray")

from drone import Drone
# from prep_routes import Horizontal
from utils import Directions

dron = Drone(
    battery=200,
    battery_usage = 0.1,
    photo_radius= 3,
    overlap= 0.5,
    move_method= "h",
    wind_direction= Directions.LEFT,
    map_size= 50,
    wind_start= 30,
    wind_stop= 40
)
dron.move(Directions.UP)
dron.move(Directions.UP)
dron.move(Directions.UP)
dron.move(Directions.UP)
dron.move(Directions.UP)
dron.move(Directions.UP)

for x in range(40):
    dirs = [Directions.UP, Directions.RIGHT]
    dron.move(random.choice(dirs))
    ax1.plot(dron.x, dron.y, marker="o", color="red")
    #print(dron)

ax2 = ax1

plt.show()
```
